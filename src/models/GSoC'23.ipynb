```python
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-es")

model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-es")
```

```python
def translate(sample_text):
  batch = tokenizer([sample_text], return_tensors='pt')
  generated_ids = model.generate(**batch)
  return tokenizer.batch_decode(generated_ids, skip_special_tokens=True)[0]
```

```python
!wget https://raw.githubusercontent.com/monum/311-translation/main/data/final_grievances.csv
```

```python
import pandas as pd
df = pd.read_csv('final_grievances.csv')
df.head()
```

```python
df['translated_text'] = df['grievance'].apply(lambda text: translate(text))
```
